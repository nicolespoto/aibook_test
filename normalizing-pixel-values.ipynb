```python
# Normalizing Pixel Values: This script demonstrates how to load an image, convert it to grayscale, and normalize its pixel values using Python.
# Step 1 Import necessary libraries
from PIL import Image
import numpy as np

# Step 2 Load an image and convert to grayscale
image = Image.open("girleshae.jpg").convert("L")  # 'L' mode = grayscale

# Step 3 Convert image to NumPy array
pixel_array = np.array(image)

# Step 4Normalize pixel values to range [0, 1]
normalized_array = pixel_array / 255.0

# Step 5 Print original and normalized pixel values for comparison
print("Original pixel value at (0,0):", pixel_array[0, 0])
print("Normalized pixel value at (0,0):", normalized_array[0, 0])
```
