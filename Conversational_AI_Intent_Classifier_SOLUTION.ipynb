```python
# Programming Exercise 8-2: Build a NLP Intent Classifier for Conversational AI using GitHub & Scikit-learn
#SOLUTION
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

texts = [
    "hello", "hi there", "good morning",
    "can you help me", "i need help", "help please",
    "bye", "goodbye", "see you later"
]

labels = [
    "greeting", "greeting", "greeting",
    "help", "help", "help",
    "bye", "bye", "bye"
]
# Part 1 TODO: Create the vectorizer
vectorizer = CountVectorizer()
# Part 2 TODO: Fit the vectorizer and transform the texts
X = vectorizer.fit_transform(texts)

# Part 3 TODO: Create and train the classifier
clf = LogisticRegression()
clf.fit(X, labels)

# Part 4 TODO: Ask the user for a message
user_message = input()
X_test = vectorizer.transform([user_message])
intent = clf.predict(X_test)[0]

# PART 5 TODO: Print the correct AI response
if intent == "greeting":
    print("AI Response: Hi! How can I assist you today?")
elif intent == "help":
    print("AI Response: Sure! Tell me what you need help with.")
elif intent == "bye":
    print("AI Response: Goodbye! Talk to you soon.")
else:
    print("AI Response: I'm not sure what you mean, but I'm here to chat!")
```
